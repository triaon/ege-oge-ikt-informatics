```python
with open('17(1).txt', 'r') as f:
    allVals = list(map(int, f.readlines()))
```

```python
results = []
for i in range(len(allVals)):
    for j in range(i+1, len(allVals)):
        if allVals[i]%160 != allVals[j]%160 and (allVals[i]*allVals[j])%7 == 0:
            results.append(allVals[i]+allVals[j])
```

```python
print(len(results), max(results))
```

```python
b = [0]
j = 0
for i in range(1, len(a)):
    if abs(a[i] - a[i-1]) <= 8:
        print('\n', i-j)
        if i - j == 1:
            b.append(b[-1] + a[i])
        else:
            b.append(a[i] + a[i-1])
        j = i
```

### Вариант учителя

```python
s = a[0]
max_s = 0

for i in range(1, len(a)):
    if abs(a[i]-a[i-1]) <= 8:
        if a[i] + s < 0:
            s = 0
        else:
            s += a[i]
    else:
        s = max(a[i],0)
    if s > max_s:
        max_s = s

print(s)
```
